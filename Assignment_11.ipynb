```python
#3.How can you create a custom exception class in Python? Provide an example that demonstrates its usage.

class CustomException(Exception):
    def __init__(self, message):
        self.message = message
        super().__init__(self.message)

# Example usage
def divide(a, b):
    if b == 0:
        raise CustomException("Division by zero is not allowed.")
    return a / b

# Handling the custom exception
try:
    result = divide(10, 0)
    print("Result:", result)
except CustomException as e:
    print("Error:", e.message)
```

#5.What is logging in Python, and why is it important in software development?


Logging in Python is a built-in module that provides a flexible and efficient way to record and store messages, warnings, and errors generated during the execution of a program. It allows developers to collect and track valuable information about the program's behavior, performance, and any issues that may arise.

Logging is important in software development for several reasons:

Debugging and Issue Investigation: When developing complex software, it's crucial to have a mechanism for capturing relevant information about the program's execution. Logging helps in identifying and diagnosing issues, tracking down bugs, and understanding the flow of execution. By logging useful data such as variable values, function calls, and stack traces, developers can gain insights into the program's behavior and pinpoint problems more efficiently.

Monitoring and Maintenance: In production environments, logging is essential for monitoring the health and performance of an application. It enables developers and system administrators to track events, errors, and warnings in real-time. Log messages can provide valuable insights into system performance, resource usage, and user behavior, aiding in performance optimization, troubleshooting, and identifying patterns of failure.

Auditing and Compliance: Logging plays a crucial role in auditing and compliance requirements. By recording relevant events and actions, logging helps maintain an audit trail for security, compliance, and legal purposes. It allows for traceability, ensuring that activities and actions can be tracked and reviewed when necessary.

Error Reporting and Alerting: Logging facilitates error reporting by capturing and documenting errors and exceptions. It allows developers to generate detailed reports with contextual information, making it easier to reproduce and fix issues. Additionally, logging can be integrated with alerting systems, notifying developers or administrators when critical errors or events occur, enabling timely response and resolution.

Performance Analysis: Logging can be used to measure and analyze the performance of an application. By recording timestamps, execution times, and other relevant metrics, developers can identify bottlenecks, optimize performance, and improve the overall efficiency of the software.

Overall, logging provides a systematic and standardized way to capture and store information about a program's execution. It promotes maintainability, troubleshooting, and optimization of software, making it an essential tool in the development and maintenance lifecycle.

```python
#6.Explain the purpose of log levels in Python logging and provide examples of when each log level would be appropriate.

In Python logging, log levels are used to categorize log messages based on their severity or importance. Each log level represents a specific level of detail or criticality, allowing developers to filter and control the verbosity of log output. Python's logging module defines several standard log levels, including:

DEBUG: The lowest log level used for detailed diagnostic information. It is typically used during development and debugging to provide extensive information about the program's flow, variable values, and function calls. Example: logging.debug("Processing data: %s", data)

INFO: A general-purpose log level used to provide informational messages that confirm the normal execution of the application. It can include high-level progress updates or important events. Example: logging.info("Server started on port %d", port)

WARNING: Used to indicate potential issues or situations that may cause problems in the future but do not prevent the program from continuing its execution. Warnings can be used to highlight unexpected scenarios or deprecated usage. Example: logging.warning("Disk space is running low")

ERROR: Represents errors that occur during the execution of the program but do not prevent it from continuing. Errors are used to indicate a problem that needs attention, such as an invalid user input or a failed operation. Example: logging.error("Failed to process file: %s", filename)

CRITICAL: The highest log level used to indicate severe errors or system failures that require immediate attention. Critical messages typically result in the termination of the program or significant disruptions. Example: logging.critical("Application crashed unexpectedly")

These log levels provide a structured approach to logging, allowing developers to prioritize and filter log messages based on their importance and severity. The appropriate log level to use depends on the specific context and purpose of the log message. During development, it's common to use lower log levels like DEBUG and INFO to gain insights into the program's behavior. In production environments, it's often recommended to set the log level to a higher value, such as WARNING, ERROR, or CRITICAL, to focus on critical issues and reduce the volume of less important log messages.

By using the appropriate log level for each log message, developers can strike a balance between providing sufficient information for troubleshooting and maintaining an appropriate level of log verbosity for efficient log analysis and monitoring.
```

```python
#7.What are log formatters in Python logging, and how can you customise the log message format using formatters?
In Python logging, log formatters are objects responsible for defining the format of log messages. They determine how the log records are rendered into text or other formats before being outputted to the desired logging destinations such as the console, files, or external services.

The logging.Formatter class is provided by the Python logging module to facilitate the customization of log message formats. This class allows developers to specify a format string that defines the structure and content of log messages. The format string can include placeholders that are replaced with actual values from the log record, such as the log level, timestamp, module name, and the log message itself.

Here's an example of customizing the log message format using a formatter:

python
Copy code
import logging

# Create a logger
logger = logging.getLogger("my_logger")
logger.setLevel(logging.DEBUG)

# Create a formatter
formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")

# Create a handler and set the formatter
handler = logging.StreamHandler()
handler.setFormatter(formatter)

# Add the handler to the logger
logger.addHandler(handler)

# Log messages
logger.debug("Debug message")
logger.info("Info message")
logger.warning("Warning message")
In the example above, we create a logger named "my_logger" and set its log level to DEBUG to ensure all log messages are captured. Then, we create a formatter using the logging.Formatter class and define a format string "%(asctime)s - %(levelname)s - %(message)s".

The format string contains placeholders enclosed in %(...) that will be replaced with actual values from the log record:

%(asctime)s: Represents the timestamp of the log record.
%(levelname)s: Represents the log level (e.g., DEBUG, INFO, WARNING, ERROR, CRITICAL).
%(message)s: Represents the log message itself.
Next, we create a StreamHandler that outputs log messages to the console. We set the formatter we created earlier as the formatter for this handler using handler.setFormatter(formatter).

Finally, we add the handler to the logger using logger.addHandler(handler).

When we log messages using the logger, the formatter will be applied, and the log messages will be displayed in the defined format:

yaml
Copy code
2021-09-01 12:34:56,789 - DEBUG - Debug message
2021-09-01 12:34:56,790 - INFO - Info message
2021-09-01 12:34:56,791 - WARNING - Warning message
By customizing the format string in the formatter, you can tailor the log message format to suit your specific requirements, including timestamps, log levels, module names, and additional contextual information. This flexibility allows for consistent and meaningful log formats that facilitate log analysis, troubleshooting, and monitoring.
```

```python
#9.What is the difference between the logging and print statements in Python? When should you use logging over print statements in a real-world application?

The logging and print statements in Python serve different purposes and have distinct characteristics. Here are the key differences between logging and print statements:

Destination: Print statements output messages directly to the standard output (console), while logging provides flexibility to direct messages to various destinations, such as files, databases, or external logging services.

Granularity: Print statements are typically used for immediate output and quick debugging during development. On the other hand, logging allows for different log levels (e.g., DEBUG, INFO, WARNING, ERROR, CRITICAL) to control the level of detail and importance of log messages.

Flexibility: Logging provides more advanced features and configurability compared to print statements. With logging, you can set different log levels, customize log message formats, use multiple loggers for different parts of your application, and enable or disable logging dynamically during runtime.

Performance Impact: Print statements can have a negative impact on performance, especially when used extensively in production code. Logging, when properly configured, can be more efficient and optimized for production environments.

In a real-world application, it is generally recommended to use logging over print statements for several reasons:

Control and Flexibility: Logging allows you to control the level of detail and importance of log messages by using different log levels. This provides more fine-grained control over the information you capture and display, making it easier to filter and manage log output.

Maintenance and Debugging: Logging provides a structured and standardized approach to capturing log messages, making it easier to maintain and debug applications. Log messages can include contextual information, timestamps, and stack traces, facilitating troubleshooting and issue investigation.

Production Environment Considerations: When running in a production environment, using print statements can clutter the console and cause performance issues. Logging allows you to direct log messages to appropriate destinations, such as log files, while keeping the console clean. It also enables you to enable or disable logging dynamically or adjust log levels without modifying the code.

Monitoring and Analysis: Logging facilitates monitoring and analysis of your application's behavior and performance. Log messages can be collected and analyzed using log management tools, enabling you to gain insights into system behavior, identify patterns, and track important events.

Overall, logging provides a more structured, configurable, and efficient approach to capturing and managing log messages in real-world applications. It offers greater control, flexibility, and scalability compared to print statements, making it a preferred choice for professional software development.
```

```python

 Here's a Python program that logs an error message to the console and a file named "errors.log" if an exception occurs during the program's execution. The error message includes the exception type and a timestamp:

python
Copy code
import logging
import traceback
import datetime

# Set up logging
logging.basicConfig(level=logging.ERROR, format='%(asctime)s - %(levelname)s - %(message)s')

# Create a file handler
file_handler = logging.FileHandler('errors.log')
file_handler.setLevel(logging.ERROR)

# Create a formatter for the file handler
file_formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
file_handler.setFormatter(file_formatter)

# Add the file handler to the logger
logging.getLogger('').addHandler(file_handler)

try:
    # Your code that may raise an exception goes here
    # For example:
    num1 = 10
    num2 = 0
    result = num1 / num2

except Exception as e:
    # Log the exception message with type and timestamp
    logging.error(f'Exception: {type(e).__name__} occurred at {datetime.datetime.now()}')
    # Log the traceback to provide additional information
    logging.error(traceback.format_exc())

    # Raise the exception again if needed for further handling
    raise
In this program, we import the logging and traceback modules. We configure logging to the ERROR level and specify a format that includes the timestamp using basicConfig().

We create a FileHandler named "errors.log" specifically for handling errors and set its level to ERROR. We also create a formatter for the file handler to format the log messages.

Next, we add the file handler to the root logger using logging.getLogger('').addHandler(file_handler) to ensure the error messages are written to the "errors.log" file.

Inside the try block, you can place your code that may raise an exception. In the example, we divide num1 by num2, intentionally causing a ZeroDivisionError.

If an exception occurs, the program will log an error message with the exception type and timestamp using logging.error(). Additionally, traceback.format_exc() is used to retrieve the full traceback information, which is also logged as part of the error message.

Finally, if necessary, you can re-raise the exception using raise to propagate the exception for further handling or program termination.
```
